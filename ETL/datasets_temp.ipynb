```python
import pandas as pd
```

```python
file_dir = '/Users/xenia/Data Bootcamp/Analysis_Projects/capstone/Resources'

athletes_data = pd.read_csv(f'{file_dir}/athletes_data.csv', low_memory=False)
print('Datasset shape:')
print(athletes_data.shape)
print(athletes_data.columns)
athletes_data.head()
```

```python
games_data = athletes_data[['Games', 'Year', 'Season', 'City']]
games_data.head(20)
```

```python
games_data_grouped = games_data.drop_duplicates()
```

```python
games_data_grouped.head()
```

```python
games_sorted = games_data_grouped.sort_values('Year')
games_sorted.head(30)
```

```python
games_sorted.reset_index(drop=True, inplace=True)
games_sorted.head()
```

```python
games_sorted['ID'] = games_sorted.index + 1
games_new_sorted = games_sorted.reindex(columns=['ID', 'Games',	'Year',	'Season',	'City'])
olympic_games = games_new_sorted
olympic_games
```

```python
olympic_games.to_csv('../Resources/olypic_games_clean.csv', index=False)
```
