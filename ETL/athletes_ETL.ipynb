```python
file_dir = '/Users/xenia/Data Bootcamp/Analysis_Projects/capstone/Resources'

athletes_data = pd.read_csv(f'{file_dir}/athletes_data.csv', low_memory=False)
print('Datasset shape:', athletes_data.shape)
print('Dataset columns:', athletes_data.columns)
athletes_data.head()
```

```python
#checking if there are NaN values in Height/Weight/Age
check_nan_in_df = athletes_data.isnull()
print (check_nan_in_df)
```

```python
athletes_data_nona = athletes_data.dropna(subset=['Age', 'Height', 'Weight', 'Sex']) #droppping Nan values in specific columns
print('No na dataset shape', athletes_data_nona.shape)
athletes_data_nona.head()
```

```python
#checking if there are NaN values in Sex/Age/Height/Weight
count_nan = athletes_data_nona['Sex'].isnull().sum()
print ('Count of NaN in athletes_data.Sex: ' + str(count_nan))
count_nan_a = athletes_data_nona['Age'].isnull().sum()
print ('Count of NaN in athletes_data.Age: ' + str(count_nan))
count_nan_h = athletes_data_nona['Height'].isnull().sum()
print ('Count of NaN in athletes_data.Height: ' + str(count_nan))
count_nan_w = athletes_data_nona['Weight'].isnull().sum()
print ('Count of NaN in athletes_data.Weight: ' + str(count_nan))
```

```python
athletes_data.dtypes
```

```python
athletes_data_nona['Age'] = athletes_data_nona['Age'].astype(int)
athletes_data_nona.head()
```

```python
#Setting new columns order
col = ['ID', 'Name', 'Sex', 'Age', 'Height', 'Weight', 'Team', 'NOC', 'Games', 'Sport', 'Event', 'Medal']
athletes_data_nona = athletes_data_nona.reindex(columns=col) 
athletes_data_nona
```

```python
athletes_data_nona['Games'].value_counts() #checking if we can drop some years
```

```python
athletes_drop_games = athletes_data_nona.groupby('Games').filter(lambda x : len(x)>70)
athletes_drop_games['Games'].value_counts()
```

```python
#Check most and less populated sports
athletes_drop_games['Sport'].value_counts()
```

```python
# it safe to assume that Rugby and Rugby Sevens are same from phisical perspective we combine those in one category

athletes_sports =  athletes_drop_games.replace('Rugby Sevens', 'Rugby')
athletes_sports['Sport'].value_counts()
```

```python
#drop Sports with insufficiant data
athletes_drop_sport = athletes_sports.groupby('Sport').filter(lambda x : len(x)>100)
athletes_drop_sport['Sport'].value_counts()
```

```python
#save clean data to csv
athletes_clean_df = athletes_drop_sport
athletes_clean_df.to_csv('../Resources/athletes_data_clean.csv', index=False)
```

# Adding column for y (1 - wining medal, 0 - no medal)

```python
athletes_data = pd.read_csv(f'{file_dir}/athletes_data.csv', low_memory=False)
print('Datasset shape:', athletes_data.shape)
print('Dataset columns:', athletes_data.columns)
athletes_data.head()

```

```python
medals = athletes_data['Medal'].to_list()

win=[]
for medal in medals:
    if pd.isna(medal):
      win.append(0)
    else:
      win.append(1)

athletes_data['Win'] = win
athletes_data
```

```python
#Adding BMI index to the dataframe - may use for visualisation
athletes_clean_df['BMI'] = athletes_clean_df.apply(lambda row: row.Weight/(row.Height*row.Height/10000) , axis=1)
athletes_clean_df.head()
```
